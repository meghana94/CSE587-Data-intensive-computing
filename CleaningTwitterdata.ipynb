```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import nltk 
import string
import re
```

```python
def load_data():
    cols = ['tweet','id']
    data = pd.read_csv('Twitterfinalsports.txt',header=None, names=cols)
    return data
```

```python
tweet_df = load_data()
tweet_df.head(100)
```

```python
df  = pd.DataFrame(tweet_df[['tweet', 'id']])
```

```python
str1 = "b'"
str2 = 'b"'
df['tweet'] = df['tweet'].apply(lambda x: re.sub(str1, '', x.lower()))
df['tweet'] = df['tweet'].apply(lambda x: re.sub(str2, '', x.lower()))
df['tweet'].head(100)
```

```python
string.punctuation
```

```python
df['tweet'] = df['tweet'].apply(lambda x:re.sub("(https?:\/\/)?([\da-z\.-]+)\.([a-z\.]{2,6})([\/\w\.-]*)*\/?\S", "",x.lower()))
df['tweet'].head(100)
```

```python
df['tweet'] = df['tweet'].apply(lambda x: re.sub('[!@#$:).;,?&_]', '', x.lower()))
df['tweet'] = df['tweet'].apply(lambda x: re.sub('[0-9]+', '', x.lower()))
df['tweet'] = df['tweet'].apply(lambda x: re.sub('  ', ' ', x))
df['tweet'].head(1000)
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

df['tweet'] = df['tweet'].apply(lambda x: remove_punct(x))
df['tweet'].head(100)
```

```python
df['tweet'] = df['tweet'].apply(lambda x: re.sub('[/x]+', '', x.lower()))
df['tweet'].head(100)
```
